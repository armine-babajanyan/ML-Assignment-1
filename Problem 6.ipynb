```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
X_nl, y_nl = make_moons(n_samples=500, noise=0.2, random_state=42)
X_train_nl, X_test_nl, y_train_nl, y_test_nl = train_test_split(X_nl, y_nl, test_size=0.2, random_state=42)

log_reg = LogisticRegression()
log_reg.fit(X_train_nl, y_train_nl)

plt.figure(figsize=(8, 6))
plt.scatter(X_test_nl[:, 0], X_test_nl[:, 1], c=y_test_nl, cmap="bwr", alpha=0.7, edgecolors="k")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Logistic Regression Decision Boundary (Non-Linear Data)")

x_min, x_max = X_test_nl[:, 0].min() - 1, X_test_nl[:, 0].max() + 1
y_min, y_max = X_test_nl[:, 1].min() - 1, X_test_nl[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))
Z = log_reg.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.3, cmap="bwr")
plt.show()
```
