```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
X = np.linspace(-3, 3, 100)
y = 0.5 * X**3 - 2 * X**2 + X + np.random.randn(100) * 3  # Cubic relationship with noise

def polynomial_features(X, degree):
    return np.vstack([X**i for i in range(degree + 1)]).T

def polynomial_regression(X, y, degree):
    X_poly = polynomial_features(X, degree)
    theta = np.linalg.inv(X_poly.T @ X_poly) @ X_poly.T @ y
    return theta, X_poly

degree = 2 
theta, X_poly = polynomial_regression(X, y, degree)

y_pred = X_poly @ theta

plt.scatter(X, y, label="Data", color="blue", alpha=0.6)
plt.plot(X, y_pred, label=f"Polynomial Regression (Degree {degree})", color="red")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.title("Polynomial Regression Fit")
plt.show()
```
